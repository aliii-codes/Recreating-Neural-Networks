# Introduction to Neural Networks

## What is a Neural Network?

**Artificial neural networks** are computational models inspired by the biological structure of the human brain. While not a perfect analogy, neural networks share key characteristics with organic brains: they consist of interconnected neurons that process information through activation patterns.

A single neuron has limited capability, but when combined with hundreds of thousands (or millions) of other neurons, the complex interconnectivity enables the network to learn patterns and make predictions that often outperform traditional machine learning methods.

### The "Black Box" Nature

Neural networks are often described as **"black boxes"** because we frequently cannot explain exactly why they reach specific conclusions. However, we do understand the mathematical mechanisms behind how they process information.

---

## Neuron Anatomy

A neuron consists of three core components:
- **Inputs**: Data fed into the neuron
- **Weights**: Multipliers applied to each input
- **Bias**: An offset value that shifts the output

The output of one neuron becomes the **input** for neurons in the next layer.

### The Core Formula

```
output = weight × input + bias
```

You can think of weights and biases as **adjustable knobs** that we tune to fit our model to the training data.

## Activation Functions

### Step Function (Basic)

For a step function, a neuron "fires" (outputs 1) if its calculated value exceeds 0, otherwise it outputs 0:

```python
output = sum(inputs × weights) + bias
```

We typically apply an **activation function** to this output:

```python
output = activation(output)
```

### Modern Activation Functions

While step functions are conceptually simple, modern neural networks use more sophisticated activation functions that provide better gradient flow during training. The most common is the **Rectified Linear Unit (ReLU)** function, which we'll explore in depth later.

Each neuron's output can either:
- Be part of the final output layer (predictions)
- Serve as input to neurons in the next layer

## Neural Network Architecture

### Example Structure

```
Input Layer        Hidden Layer 1     Hidden Layer 2     Output Layer
(3 neurons)        (4 neurons)        (4 neurons)        (1 neuron)

x₁  ────┐       ○ ReLU ───┐        ○ ReLU ───┐
├──→   ○ ReLU    ├──→     ○ ReLU    ├──→   ŷ
x₂  ────┤       ○ ReLU   ┤        ○ ReLU   │
│       ○ ReLU ──┘        ○ ReLU ──┘
x₃  ────┘

Each connection: output = sum(inputs × weights) + bias → ReLU activation
```

### Layer Types

- **Input Layer**: Receives raw or preprocessed numeric data
- **Hidden Layers**: Transform data through weighted connections and activations
- **Output Layer**: Produces predictions (one neuron per class for classification, or single neuron for regression)
