# A Single Neuron 
``` python
# Implementig a Neural Network

inputs = [1,2,3]
weights = [0.2,0.4,0.6]
bias = 2

output = (inputs[0]*weights[0]+  # Multiply 1 with 0.2
          inputs[1]*weights[1]+  # Multiply 2 with 0.4
          inputs[2]*weights[2]+  # Multiply 3 with 0.6
          bias)                   # Add bias

print(output)
```



````python
# Implementig a Neural Network

inputs = [1,2,3]
weights = [0.2,0.4,0.6]
bias = 2

output = (inputs[0]*weights[0]+  # Multiply 1 with 0.2
          inputs[1]*weights[1]+  # Multiply 2 with 0.4
          inputs[2]*weights[2]+  # Multiply 3 with 0.6
          bias)                   # Add bias

print(output)
````

# A Layer of Neurons 
``` python 
inputs = [1,2,3,2.5]
weights1 = [0.2, 0.4, 0.6, 0.5]
weights2 = [0.4, -0.5, 0.6, 0.5]
weights3 = [-0.3, 0.7, 0.1, -0.8]
bias1 = 2
bias2 = 3
bias3 = 0.5


output = (# Neuron 1
          inputs[0]*weights1[0]+
          inputs[1]*weights1[1]+
          inputs[2]*weights1[2]+
          inputs[3]*weights1[3]+
          bias1,
          # Neuron 2
          inputs[0]*weights2[0]+
          inputs[1]*weights2[1]+
          inputs[2]*weights2[2]+
          inputs[3]*weights2[3]+
          bias2,
          # Neuron 3
          inputs[0]*weights3[0]+
          inputs[1]*weights3[1]+
          inputs[2]*weights3[2]+
          inputs[3]*weights3[3]+
          bias3)
print(output)
```

````python
inputs = [1,2,3,2.5]
weights1 = [0.2, 0.4, 0.6, 0.5]
weights2 = [0.4, -0.5, 0.6, 0.5]
weights3 = [-0.3, 0.7, 0.1, -0.8]
bias1 = 2
bias2 = 3
bias3 = 0.5


output = (# Neuron 1
          inputs[0]*weights1[0]+
          inputs[1]*weights1[1]+
          inputs[2]*weights1[2]+
          inputs[3]*weights1[3]+
          bias1,
          # Neuron 2
          inputs[0]*weights2[0]+
          inputs[1]*weights2[1]+
          inputs[2]*weights2[2]+
          inputs[3]*weights2[3]+
          bias2,
          # Neuron 3
          inputs[0]*weights3[0]+
          inputs[1]*weights3[1]+
          inputs[2]*weights3[2]+
          inputs[3]*weights3[3]+
          bias3)
print(output)
````
