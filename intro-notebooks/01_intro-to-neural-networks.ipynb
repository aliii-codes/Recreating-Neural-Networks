# What is a Neural Network?

**Artificial** neural networks are inspired by the organic brain, translated to the computer. It's not a perfect comparison, but there are neurons, activation, and lots of interconnectivity, even if the underlying processes are quite different.

A single neuron by itself is relatively useless, but when combined with hundreds of thousands of (or many more) other neurons, the interconnectivity produces relationships and results that frequently outperform any other ML methods.

**Neural Networks** are considered to be "black boxes" in that we often have no idea why they reach the conclusions they do. We do understand how they do this, though.

A neuron consists of an input, weight, and a bias.  
Output of one neuron becomes **Input** for another.

The concept of biases and weights can be thought of as **knobs** that we can tune to fit our model to data.

**output = weight * input + bias**

Now here we can see how weights and biases impact a single neuron  
**Link:** https://nnfs.io/bru/

For a step function, if the neuron's output value, which is calculated by sum(inputs × weights) + bias, is greater than 0, the neuron fires (so it would output a 1). Otherwise, it does not fire and would pass along a 0. The formula for a single neuron might look like:

```python
output = sum(inputs * weights) + bias
```

We usually apply an activation function to this output, noted by activation()
```python
output = activation(output)
```

While you can use a step function for your activation function, we tend to use slightly more advanced. Neural Networks of today tend to use more informative activation functions (rather than a step function), such as the **Rectified Linear** (ReLU) activation function, which we will cover soon in depth. Each neuron's output could be a part of the ending output layer, as well as the input to another layer of neurons. While the full activation of a neural network can get very large, let's start with a simple example with 2 hidden layers of 4 neurons each.

## Neural Network Structure

``` 
Input Layer        Hidden Layer 1     Hidden Layer 2     Output Layer
(3 neurons)        (4 neurons)        (4 neurons)        (1 neuron)
x₁  ────┐       ○ ReLU ───┐        ○ ReLU ───┐
├──→   ○ ReLU    ├──→     ○ ReLU    ├──→   ŷ
x₂  ────┤       ○ ReLU   ┤        ○ ReLU   ┤
│       ○ ReLU ──┘        ○ ReLU ──┘
x₃  ────┘
Each connection: output = sum(inputs × weights) + bias → ReLU activation
```

> Input layer = raw/preprocessed numeric data. Output layer = predictions (one neuron per class, or single neuron for binary).

Here **https://nnfs.io/qtb** you can see the passing of image data through a neural network, getting a classification.
