# A Single Neuron

Let's implement a basic neural network with a single neuron. The neuron takes inputs, applies weights, adds a bias, and produces an output.

````julia
inputs = [1, 2, 3]
weights = [0.2, 0.4, 0.6]
bias = 2

output = (inputs[0] * weights[0] +  # Multiply 1 with 0.2
          inputs[1] * weights[1] +  # Multiply 2 with 0.4
          inputs[2] * weights[2] +  # Multiply 3 with 0.6
          bias)                    # Add bias

print(output)
````

# A Layer of Neurons

Now let's extend this to a layer with multiple neurons. Each neuron has its own set of weights and bias.

````julia
inputs = [1, 2, 3, 2.5]
weights1 = [0.2, 0.4, 0.6, 0.5]
weights2 = [0.4, -0.5, 0.6, 0.5]
weights3 = [-0.3, 0.7, 0.1, -0.8]
bias1 = 2
bias2 = 3
bias3 = 0.5

output = (
    # Neuron 1
    inputs[0] * weights1[0] +
    inputs[1] * weights1[1] +
    inputs[2] * weights1[2] +
    inputs[3] * weights1[3] +
    bias1,
    # Neuron 2
    inputs[0] * weights2[0] +
    inputs[1] * weights2[1] +
    inputs[2] * weights2[2] +
    inputs[3] * weights2[3] +
    bias2,
    # Neuron 3
    inputs[0] * weights3[0] +
    inputs[1] * weights3[1] +
    inputs[2] * weights3[2] +
    inputs[3] * weights3[3] +
    bias3
)

print(output)
````

# Automating with Loops

Writing out each calculation manually becomes impractical as we scale to hundreds or thousands of neurons. Let's use loops to automate this process.

````julia
inputs = [1, 2, 3, 2.5]

# Each inner list represents one neuron's weights (one weight per input)
weights = [[0.2, 0.8, -0.5, 1],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]

# One bias per neuron
biases = [2, 3, 0.5]

layer_outputs = []

# Loop through each neuron (its weights + bias)
for neuron_weights, neuron_bias in zip(weights, biases):
    neuron_output = 0

    # Multiply each input by its corresponding weight and accumulate
    for n_input, weight in zip(inputs, neuron_weights):
        neuron_output += n_input * weight  # sum(inputs * weights)

    neuron_output += neuron_bias  # Add bias
    layer_outputs.append(neuron_output)  # Store this neuron's output

print(layer_outputs)
````

# Tensors, Arrays, and Vectors

## What are Tensors?

**Tensors** are closely related to arrays. In machine learning contexts, people often use the terms "tensor," "array," and "matrix" interchangeably without much issue. However, there are subtle differences, primarily in context or the attributes of the tensor object.

### Python Lists

To understand tensors, let's first review Python's data containers:

**A single list:**
```python
l = [1, 2, 3]
```

**A list of lists (2D):**
```python
lol = [[1, 2, 3, 5],
       [3, 2, 1, 3]]
```

**A list of lists of lists (3D):**
```python
lolol = [[[1, 5, 6, 2],
          [3, 2, 1, 3]],
         [[5, 2, 1, 2],
          [6, 4, 8, 4]],
         [[2, 8, 5, 3],
          [1, 1, 9, 4]]]
```

### The Definition

In computer science and deep learning, extensive debate exists about the precise definition of tensors versus arrays. This debate often stems from people arguing from different perspectives. 

The key takeaway is:

**A tensor object is an object that can be represented as an array.**

For practical purposes in deep learning, you can think of tensors as multi-dimensional arrays that support efficient mathematical operations.

# Dot Product and Vector Addition


A dot product of two vectors is a sum of products of consecutive vector elements. Both vectors must be of the same size (hav an equal number of elements)
<br>

**Formula**
``` bash
dot(A,B) = sum of (A[i] * B[i]) for every i 
```

or more easily it's :

$$\vec{a} \cdot \vec{b} = \sum_{i=1}^{n} a_i b_i = a_1b_1 + a_2b_2 + ... + a_nb_n$$

**Lets now implement everything (Dot product) in Code**


````julia
a = [1,2,3,4]
b = [3,4,5,6]

dot = sum(x * y for x, y in zip(a, b))
print(dot)
````

# Lets just confirm it 

````julia
import numpy as np

a = [1,2,3,4]
b = [3,4,5,6]

dot = np.dot(a,b)
print(dot)
````

# A Single Neuron in Numpy

Let's code the solution for a single neuron to start , using the dot product and the addition of the vectors in NumPy.

````julia
import numpy as np

inputs = [1, 2, 3, 2.5]
weights = [0.2, 0.8, -0.5, 1]
bias = 2.0

output = np.dot(weights, inputs) + bias
print(output)
````

# A layer of neurons with Numpy

````julia
import numpy as np

inputs = [1, 2, 3, 2.5]
weights = [[0.2, 0.8, -0.5, 1],
          [0.5, -0.91, 0.26, -0.5],
          [-0.26, -0.27, 0.17, 0.87]]
biases = [2, 3, 0.5]

layer_outputs = np.dot(weights, inputs) + biases

print(layer_outputs)
````

# Martix Product

The **matrix product** is an operation in which we have 2 matrices, and we are performing dot products of all combinations of rows from the first matrix and the columns of the 2nd matrix , resulting in a matrix of those atomic **dot products**:
```
A (2×3)          B (3×2)          Output (2×2)

[ a b c ]   ×   [ d e ]   =   [ · · ]
[ f g h ]       [ i j ]       [ · · ]
                [ k l ]
```

# Transposition for the Matrix Product 

How did we suddenly go from 2 vectors to row and column vectors? We used the relation of the 
dot product and matrix product saying that a dot product of two vectors equals a matrix product of 
a row and column vector (the arrows above the letters signify that they are vectors):

$$\vec{a} \cdot \vec{b} = \vec{a}^T \vec{b}$$

We also have temporarily used some simplification, not showing that column vector b is actually 
a transposed vector b. The proper equation, matching the dot product of vectors a and b written 
as matrix product should look like:

$$\vec{a} \cdot \vec{b} = \vec{a}^T (\vec{b}^T)^T$$
