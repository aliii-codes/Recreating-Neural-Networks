# Activation Functions

Activation functions are a crucial component of neural networks. They introduce non-linearity into the network, allowing it to learn complex patterns and make decisions.

## Why Do We Need Activation Functions?

Without activation functions, a neural network would simply be a linear regression model, no matter how many layers it has. The activation function allows the network to approximate any continuous function, making it a universal function approximator.

## Common Activation Functions

### 1. ReLU (Rectified Linear Unit)

The most widely used activation function in modern deep learning.

**Formula:**
$$
f(x) = \max(0, x)
$$

**Characteristics:**
- Outputs 0 for negative inputs
- Outputs the input value for positive inputs
- Computationally efficient
- Helps mitigate the vanishing gradient problem

### 2. Sigmoid

Often used in the output layer for binary classification.

**Formula:**
$$
f(x) = \frac{1}{1 + e^{-x}}
$$

**Characteristics:**
- Outputs values between 0 and 1
- Smooth gradient
- Can suffer from vanishing gradient problem

### 3. Softmax

Used in the output layer for multi-class classification.

**Formula:**
$$
f(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
$$

**Characteristics:**
- Outputs a probability distribution
- All outputs sum to 1
- Used for classification problems
